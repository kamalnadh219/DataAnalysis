## Daily Returns For a Stock

```python
stocks['Daily_Return']=stocks['Close'].pct_change()
```

```python
stocks
```

```python
plt.hist(stocks['Daily_Return'],bins=50,color='red',label='Daily return')
plt.xlabel("Daily_return")
plt.ylabel("Count")
plt.title('Distribution of Daily Return')
plt.show()
```

### Identifying Corelation betweening variables 

```python
sns.heatmap(stocks.corr())
```

### Periods with Highest Volatility

```python
rolling_volatility = stocks['Daily_Return'].rolling(window=30).std()
plt.figure(figsize=(12, 8))
plt.plot(stocks['Date'], rolling_volatility, label='30-Day Rolling Volatility')
plt.title('Stock Volatility Over Time')
plt.xlabel('Date')
plt.ylabel('Volatility')
plt.legend()
plt.show()
```

### Average Volatility

### Trading Volume Variations Across Months

### Short-term and Long-term Moving Averages

```python
stocks
```

```python
stocks['short_MA']=stocks['Close'].rolling(20).mean()
```

```python
stocks['long_MA']=stocks['Close'].rolling(100).mean()
```

```python
plt.figure(figsize=(12, 8))

plt.plot(stocks['Date'], stocks['short_MA'], label='20-Day MA')
plt.plot(stocks['Date'], stocks['long_MA'], label='100-Day MA')
plt.title('Short-term and Long-term Moving Averages')
plt.xlabel('Date')
plt.ylabel('Price')
plt.legend()
plt.show()
```

### Specific Days or Months with High Trading Volumes

```python
high_volume_days = stocks[stocks['Volume'] > stocks['Volume'].quantile(0.95)]
high_volume_days[['day','month']]
```

### Outlier detection

```python
price_outliers = stocks[(np.abs(stocks['Close'] - stocks['Close'].mean()) > (3 * stocks['Close'].std()))]
volume_outliers = stocks[(np.abs(stocks['Volume'] - stocks['Volume'].mean()) > (3 * stocks['Volume'].std()))]
```
